HW (ให้ใช้เฉพาะที่อาจารย์สอนไปแล้วในวิชานี้)

1.   คำนวณ อายุเฉลี่ย ของผู้หญิง และผู้ชาย ของข้อมูลทั้งหมด
2.   คำนวณ อายุเฉลี่ย ของผู้หญิง และผู้ชาย ของผู้ป่วยในจังหวัดขอนแก่น

3. หาจำนวนผู้ป่วยที่เป็นคน "**คนต่างชาติเดินทางมาจากต่างประเทศ**"

````
this_data = data_covid[['sex','age','province_of_onset']]
````

````
this_data
````

**ผู้ป่วยหญิง**

````
female = this_data[this_data['sex']=='หญิง']
female['age']
````

````
#เป็นการแปลงข้อมูลในคอลัมน์ 'age' ของ DataFrame female ให้เป็นตัวเลขครับ
female['age'] = pd.to_numeric(female['age'], errors='coerce')
````

`pd.to_numeric()`: เป็นฟังก์ชันของไลบรารี pandas ที่ใช้ในการแปลงข้อมูลให้เป็นชนิดตัวเลข

`female['age']`: คือการเลือกคอลัมน์ 'age' จาก DataFrame female

`errors='coerce'`: เป็นอาร์กิวเมนต์ที่สำคัญ หมายความว่า หากพบค่าที่ไม่สามารถแปลงเป็นตัวเลขได้ (เช่น เป็นข้อความที่ไม่ใช่ตัวเลข) ให้แทนที่ค่านั้นด้วย NaN (Not a Number) แทนที่จะแสดงข้อผิดพลาด (error)

````
sum_age = 0
N = 0
for a in female['age']:
    if not pd.isna(a) and a > 0:
        sum_age += a
        N += 1

print(f'อายุเฉลี่ย ของ ผู้ป่วยหญิง {sum_age/N}')
````

````
female_KK = female[female['province_of_onset']=='ขอนแก่น']

sum = 0
N = 0
for a in female_KK['age']:

    if a > 0:
        sum += a # sum = sum + a
        N += 1

print(f'อายุเฉลี่ย ของ ผู้ป่วยหญิง ในจังหวัดขอนแก่น {sum/N}')
````

## Quiz ตัดตารางออกมาเฉพาะปี 2021 ```announce_date``` ในปี 2021

**Hint**
*   วนลูปหา index ของปี 2021
*   ตัดตารางมาเฉพาะ ปี 2021



````
data_covid
````

````
# สร้าง list เปล่าสำหรับเก็บค่า True/False เพื่อใช้ในการกรองข้อมูล
TF = list()
# วนลูปผ่านแต่ละแถวใน DataFrame 'data_covid'
for each_row in data_covid.iterrows():
    # แยกสตริงวันที่ในคอลัมน์ 'announce_date' ด้วยเครื่องหมาย '/' และเลือกส่วนที่เป็นปี (index 2)
    # ตรวจสอบว่าปีที่ได้มาเท่ากับ '2021' หรือไม่
    if each_row[1]['announce_date'].split('/')[2] == '2021':
      # ถ้าเป็นปี 2021 ให้เพิ่ม True ลงในลิสต์ TF
      TF.append(True)
    else:
      # ถ้าไม่ใช่ปี 2021 ให้เพิ่ม False ลงในลิสต์ TF
      TF.append(False)

# ใช้ลิสต์ TF (Boolean Series) ในการกรอง DataFrame 'data_covid' และแสดงผล 5 แถวแรก
data_covid[TF].head()
````

````
TF=list()
for each_row in data_covid.iterrows():
    if each_row[1]['announce_date'].split('/')[2] == '2021':
        TF.append(True)
    else:
        TF.append(False)

data_covid[TF].head()
````
